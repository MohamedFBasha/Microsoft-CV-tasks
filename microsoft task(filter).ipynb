```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
def apply_filter(kernel_size):
   
    img = cv2.imread('E:/my photo/wall paper/AI.png')

    
    averaged_image = cv2.blur(img, (kernel_size, kernel_size))

   
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('the first image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(averaged_image, cv2.COLOR_BGR2RGB))
    plt.title('Averaged Image')
    plt.axis('off')

    plt.show()

interact(apply_filter, kernel_size=widgets.IntSlider(min=3, max=15, step=2, value=5))
```
